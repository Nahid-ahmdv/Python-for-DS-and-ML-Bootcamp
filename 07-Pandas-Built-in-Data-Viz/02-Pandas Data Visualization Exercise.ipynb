## Bonus Challenge!
Note, you may find this really hard, reference the solutions if you can't figure it out!
**Notice how the legend in our previous figure overlapped some of actual diagram. Can you figure out how to display the legend outside of the plot as shown below?**

**Try searching Google for a good stackoverflow link on this topic. If you can't find it on your own - [use this one for a hint.](http://stackoverflow.com/questions/23556153/how-to-put-legend-outside-the-plot-with-pandas)**

```python
ax = df3.iloc[:31].plot.area(alpha=0.4)
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.ylim(0, 3);
'''
loc='center left': aligns the legend to the center left of the bounding box.
bbox_to_anchor=(1.0, 0.5): places the bounding box outside the plot at the middle right (1.0 is the right edge of the plot, 0.5 is vertical center).
'''
```

```python
f = plt.figure()
df3.iloc[0:31].plot.area(alpha=0.4,ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
'''
the part ax=f.gca() means:

 f.gca():
gca stands for "get current axes".

f.gca() gets the current Axes object from the Figure object f.

This tells Pandas to plot directly on the axes of the figure f.

 Why use it?
When you create a figure manually with f = plt.figure(), it doesnt automatically come with an axes.

f.gca() makes sure your Pandas plot is drawn inside that specific figure.

Summary:
You're telling Pandas:

Plot on the axes of figure f, not on a new figure.
''';
```
