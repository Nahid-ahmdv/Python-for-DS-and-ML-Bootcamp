```python
sns.histplot(data=tips, x='total_bill', kde=True, stat='density', bins=20, color='skyblue', edgecolor='blue');
'''
âœ… Explanation:
data=tips:
Specifies the DataFrame (tips) that contains your data.

x='total_bill':
Chooses the 'total_bill' column from the DataFrame to plot on the x-axis.

kde=True:
Adds a KDE (Kernel Density Estimate) line on top of the histogram. This shows the smoothed probability density curve â€” a continuous estimate of the distribution.

stat='density':
Normalizes the histogram so that the total area under the bars equals 1, making it compatible with the KDE curve. Without this, the KDE and histogram would be on different scales.

bins=20:
Sets the number of bars (bins) in the histogram to 20. More bins = more granularity.

color='skyblue':
Fills the bars with a sky blue color.

edgecolor='blue':
Outlines each bar with a blue border, making them visually distinct.


ðŸ§  What is Kernel Density Estimate (KDE)?
KDE stands for Kernel Density Estimate. It's a smooth curve that estimates the probability distribution of a dataset â€” basically, it shows where the data is concentrated.

ðŸ” In simple terms:
Imagine taking each data point and placing a little bell-shaped curve (called a kernel, usually a Gaussian) over it. Then you add all those little curves together to get a smooth estimate of the full distribution.

ðŸ“Š Histogram vs KDE:
A histogram breaks data into bins and shows how many values fall into each bin (discrete).

A KDE smooths out the histogram into a continuous curve â€” giving a clearer view of distribution shape without depending on bin size.

ðŸ“ˆ Why use KDE?
To visualize the shape of the data (e.g., is it unimodal, bimodal? skewed?).

To get a more flexible and less noisy view than a histogram, especially with smaller datasets.

ðŸ“˜ Why it's called a kernel:
A kernel is a weighting function â€” it gives more weight to points near a target and less weight to points farther away. In KDE, it's used to estimate the distribution by placing a smooth, symmetric function (like a bell curve) on each data point.

Think of the kernel as the "core influence" of a data point on the surrounding space â€” hence the word â€œkernel,â€ which literally means "core" or "center."

ðŸŽ¯ In KDE:
A kernel function (e.g., Gaussian, Epanechnikov, or Uniform) determines the shape of the little curve centered at each data point.

These kernels are then added together to produce the smooth overall curve you see in the plot.
''';

```

```python
sns.distplot(tips['total_bill'], kde=True, hist=True,
             hist_kws={'color': 'skyblue', 'edgecolor': 'blue'},
             kde_kws={'linewidth': 2});
'''
tips['total_bill']: same data as x='total_bill'

bins=20: number of histogram bins

kde=True: show KDE curve

hist=True: show histogram

hist_kws={'color': ..., 'edgecolor': ...}: matches color and edgecolor in histplot

In kde_kws={'linewidth': 2}, the term kws stands for "keyword arguments". It make the KDE line thicker with a linewidth of 2 (default is usually 1).

stat='density': distplot defaults to density when kde=True, so it matches automatically

âš ï¸ Note: sns.distplot() is deprecated in recent versions of Seaborn (0.11+), so it's best to use sns.histplot() or sns.kdeplot() going forward.
''';
```

## jointplot

jointplot() allows you to basically match up two distplots (histplots or displots) for bivariate data. Seabornâ€™s jointplot() is a powerful tool used for visualizing the relationship between two variables (i.e., bivariate data). It creates a main plot showing how the variables relate (like a scatterplot), and it also includes two smaller histograms (or KDEs) on the top and right sides that show the distribution of each variable individuallyâ€”like two distplots. With your choice of what **kind** parameter to compare with.
The **kind** argument in jointplot() lets you choose how you want to visualize the relationship between the two variables. Here are the options:

* â€œscatterâ€  (default): A basic scatter plot showing each point.

* â€œregâ€ : A scatter plot with a regression line fit to the data (plus confidence intervals).

* â€œresidâ€: Shows the residuals (differences between actual and predicted values from regression).

* â€œkdeâ€ : Uses Kernel Density Estimation to show the smoothed density of the two variables together.

* â€œhexâ€: A hexbin plot, which bins points into hexagonal tiles and colors them based on how many points fall into each tile (great for large datasets).

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg');
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
import pandas as pd

# Load the dataset
tips = sns.load_dataset("tips")

# Calculate regression parameters
slope, intercept, r_value, p_value, std_err = linregress(tips['total_bill'], tips['tip'])

# Print regression equation
print(f"Regression equation: tip = {slope:.2f} * total_bill + {intercept:.2f}")
print(f"The correlation coefficient: {r_value:.2f}")
print(f"p_value: {p_value:.4f}")
# Plot using seaborn
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
plt.show()
'''
ðŸ“ˆ slope
The coefficient of x in the regression line.

It tells you how much y changes for every 1-unit increase in x.

In the example:
tip = 0.11 * total_bill + 0.92
The slope is 0.11, meaning: for every $1 increase in the total bill, the tip increases by $0.11 (on average).

ðŸ“‰ intercept
The value of y when x is 0.

It represents the starting point of the line on the y-axis.

In the example, intercept = 0.92, so if the total bill were $0, the predicted tip would be $0.92 (theoretically, though not realistic in context).

ðŸ“Š r_value
The correlation coefficient (r).

It measures how strong the linear relationship is between x and y.

Ranges from:

+1 = perfect positive correlation

-1 = perfect negative correlation

0 = no correlation

The square of this (r_valueÂ²) is R-squared, which tells you the proportion of variance in y explained by x.

ðŸ” p_value
Tests the null hypothesis that the slope is zero (no relationship).

A small p-value (typically < 0.05) means the relationship is statistically significant.

If p_value < 0.05, we usually say the slope is not due to random chance.

âš ï¸ std_err (Standard Error of the Slope)
Measures the uncertainty or variability in the slope estimate.

Smaller values = more reliable estimate of the slope.

It helps in creating confidence intervals for the slope.
''';
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde');
'''
"Smoothed density" refers to a continuous estimate of how the data is distributed, instead of showing the raw, binned values (like in a histogram).
Let me explain with a comparison:

ðŸ”¹ Histogram:
It splits the data into bins (e.g. 0â€“10, 10â€“20, etc.).

Then it counts how many data points fall in each bin.

It's discreteâ€”you get bars.

ðŸ”¹ Smoothed Density (via KDE - Kernel Density Estimate):
Instead of counting bins, it places a smooth curve (called a kernel) over each data point.

Then it adds up all these curves to get a smooth continuous curve.

The result is a line that estimates the probability density of the data.

Example:
If you have a histogram that looks like a jagged mountain range, KDE smooths it out into a rolling hill â€” helping you see patterns and shape more clearly (like whether the distribution is skewed or has multiple peaks).


'''

'''
The plot below is a Seaborn jointplot() using the kind='kde' option, which shows a bivariate Kernel Density Estimate (KDE).

ðŸ” What You're Seeing:
1. Main Plot (Center):
This is a contour plot, like a topographic map.

Each contour line encloses a region with higher density.

The closer the lines, the higher the concentration of data points.

This tells you where the data is most densely packed (in this case, where combinations of total_bill and tip values are most frequent).

2. Top and Right Plots:
These are univariate KDE plots.

Top shows the smoothed distribution of total_bill.

Right shows the smoothed distribution of tip.

These help you see the spread and shape of each individual variable.

ðŸ§  Interpretation:
The densest area (most data) is where total_bill is around 10â€“20 and tip is around 2â€“4.

The shape of the contours shows that as the bill increases, tips tend to increase, but the density becomes more spread out (less frequent for higher values).
'''
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid');
'''
sns.jointplot(x='total_bill', y='tip', data=tips, kind='resid') creates a residual plot using Seabornâ€™s jointplot.

Explanation:
- x='total_bill': Independent variable (predictor).
- y='tip': Dependent variable (response).
- kind='resid': Specifies that the plot should be a residual plot.

What the residual plot does:
1. Fits a linear regression model predicting 'tip' from 'total_bill'.
2. Calculates residuals for each data point:
      Residual = y_i - Å·_i
      where:
        y_i  = actual tip value,
        Å·_i = predicted tip value from the regression line.
3. Plots these residuals on the y-axis against 'total_bill' on the x-axis.

Interpretation:
- Residuals scattered randomly around zero indicate a good linear fit.
- Patterns or systematic structures in residuals suggest issues such as:
    * Non-linearity,
    * Heteroscedasticity (changing variance),
    * Model misspecification.

This plot is useful for diagnosing how well the linear model fits the data.

'''
```

### Comparison: `sns.pairplot(tips)` vs `pandas.plotting.scatter_matrix(tips)`

Both functions are used to visualize pairwise relationships between numeric columns in a dataset, but **Seaborn's `pairplot`** is more powerful and customizable.

**âœ… Similarities**
- Both show a grid of scatter plots for each pair of numeric variables.
- Diagonal plots typically show histograms or KDEs of each variable.
- Useful for detecting correlations, clusters, and patterns.

**ðŸ”¹ Differences**

| Feature                 | `sns.pairplot()`                      | `pd.plotting.scatter_matrix()`       |
|-------------------------|-------------------------------------|-------------------------------------|
| **Style**               | Seaborn style (modern, attractive)  | Matplotlib default (plainer)         |
| **Diagonal Options**     | KDE or histogram                    | Histogram only                      |
| **Color Grouping (hue)** | âœ… Supports hue for category coloring | âŒ Not supported                    |
| **Plot Customization**   | Easier, more aesthetic              | Less flexible                      |
| **Built-in Regression**  | Can add regressions via `kind='reg'` | âŒ Not built-in                    |


**Summary:**  
`sns.pairplot()` offers a more visually appealing and flexible way to explore pairwise relationships, especially when you want to include categorical coloring or customize plots. Meanwhile, `scatter_matrix()` is a simpler, more basic alternative available directly in pandas.

## Understanding Confidence Intervals (CI) in Regression Plots

The confidence interval (CI) shown in your regression plot represents the uncertainty around the **true regression line**, not the predicted values for individual data points. It's important to distinguish this from a prediction interval (PI).

---

### Confidence Intervals (CIs) in Your Plot

In your plot, the confidence intervals (CIs) appear as the **shaded regions around the regression lines** in the scatter plots.

- The **solid line** in each scatter plot shows the best-fit linear regression line for that gender group.
- The **shaded area** around the line is the confidence interval, visually representing uncertainty about where the true regression line lies.
- A **narrower CI** means more confidence in the estimated relationship between the two variables.
- A **wider CI** suggests more variability or fewer data points, meaning less certainty about the exact relationship.

#### Interpretation Example:  
In the *tip vs. total_bill* plot:

- Both males and females show a positive linear relationship.
- The blue (Male) and red (Female) shaded areas around the regression lines indicate that we are **95% confident** the true regression line for each group lies within those bounds.

---

### Confidence Interval (CI) vs. Prediction Interval (PI)

| Aspect                     | Confidence Interval (CI)                             | Prediction Interval (PI)                          |
|----------------------------|-----------------------------------------------------|-------------------------------------------------|
| **What it estimates**      | Where the *true regression line* is likely to lie  | Where *individual new data points* are likely to fall |
| **Visual representation** | Shaded band around the regression line              | Not shown by default; requires special calculation |
| **Width**                  | Narrower, reflects uncertainty in average trend     | Wider, accounts for both line uncertainty and natural scatter |
| **Interpretation**         | Confidence about the *average trend* (the line)     | Confidence about *single predictions* (individual points) |

---

### Sources of Uncertainty in the Regression Line (and CI width)

The width of the confidence interval depends on several key factors related to your data and model:

#### 1. Sampling Variability
- Models are built from a sample, not the entire population.
- Different samples would produce slightly different regression lines.
- The CI reflects the range of lines you'd expect if the study were repeated many times.

#### 2. Scatter in the Data (Residuals)
- More scatter (noise) around the regression line increases uncertainty.
- Less scatter â†’ tighter CI; more scatter â†’ wider CI.

#### 3. Number of Data Points
- Fewer points â†’ more uncertainty.
- More points â†’ stronger estimates â†’ narrower CI.

#### 4. Leverage (Distribution of X-values)
- Points far from the mean of the predictor variable (e.g., `total_bill`) have higher leverage.
- Predictions at these points have greater uncertainty.
- This causes CIs to widen at the ends of the regression line.

## kdeplot

kdeplots are [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value. For example:

```python
'''
This script is a manual visualization of Kernel Density Estimation (KDE), built from scratch using NumPy, 
Matplotlib, and SciPy. It demonstrates how KDE works by plotting individual Gaussian kernels (aka "bumps") for each data point.
'''

# Don't worry about understanding this code!
# It's just for the diagram below
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#Create dataset
dataset = np.random.randn(25)   #Generates 25 samples from a standard normal distribution.
# dataset = tips['total_bill']
# Create another rugplot
sns.rugplot(dataset);   #Places small ticks for each data point along the x-axis â€” gives you a feel for where the data is.

# Set up the x-axis for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min,x_max,100)   #Defines 100 points between min - 2 and max + 2 to evaluate the KDE curve and individual Gaussians on a smooth axis.

# Set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'
# Estimate Bandwidth (the key step)
bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2
'''
This is Silverman's Rule of Thumb:  h = \\left(\\frac{4\\sigma^5}{3n}\\right)^{1/5}
It estimates the standard deviation (spread) for each Gaussian centered at each data point.
'''

# Create an empty kernel list
kernel_list = []

# Plot each basis function (Manually build and plot each Gaussian kernel)
for data_point in dataset:
    
    # Create a kernel for each point and append to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    '''
    For each data point, a normal distribution PDF is computed using:
        center = the data point
        std (spread) = bandwidth
    1. stats.norm(data_point, bandwidth):This creates a normal (Gaussian) distribution object from SciPy's stats.norm, with:
        mean = data_point
        standard deviation = bandwidth
    So now you have a frozen normal distribution centered at your data point. Think of this like saying:
    "Make me a bell curve centered at this data point, with the spread of bandwidth."
    But this alone just defines the distribution â€” it doesnâ€™t give you the actual values yet. 
    2. .pdf(x_axis): This then calls the Probability Density Function (PDF) of the Gaussian distribution at every x value in x_axis. It means:
    â€œEvaluate the height of a Gaussian bell curve (centered at data_point, width = bandwidth) for each value in x_axis.â€
    If stats.norm(mean, std) gives you a car, then .pdf(x) is like driving that car on a specific road (x-values) to see how high the curve is at each point.
    '''
    kernel_list.append(kernel)
    
    #Scale for plotting (these lines are purely for visualization and help you scale the Gaussian curves so they fit nicely in the plot.)
    kernel = kernel / kernel.max()
    '''
    This normalizes the kernel â€” i.e., it rescales it so that the maximum height becomes 1.
    Why? The raw values from .pdf(x) can vary depending on:
        -The bandwidth (std of the Gaussian)
        -Where the data point is
    By dividing by kernel.max(), you ensure every curve peaks at 1, regardless of how wide or narrow it is.
    So now each Gaussian bump will have the same visual height in the plot (before further scaling).
    '''
    kernel = kernel * .4
    '''
    After normalization, this line shrinks the height of each bump to 0.4 units tall (instead of 1).
    Why 0.4? This number is arbitrary â€” chosen to make the bumps visually clear without overlapping too much.
    It also ensures they stay within a reasonable y-range (since the plot sets plt.ylim(0, 1)).
    You could use 0.2, 0.6, or any number that looks good visually â€” it does not affect the logic of KDE itself.
    '''
    #create a plot using Functional method:
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)

plt.ylim(0,1);
```

```python
#  This code block is computing and plotting the manual Kernel Density Estimation (KDE) by summing the individual Gaussian kernels (basis functions) you built earlier.


# To get the kde plot we can sum these basis functions.

# Plot the sum of the basis function
sum_of_kde = np.sum(kernel_list,axis=0)
'''
This is the core of KDE:
    -kernel_list contains one Gaussian curve for each data point.
    -np.sum(..., axis=0) adds up all those curves point by point (along the x-axis).
Imagine you had 25 y-arrays (one for each Gaussian). This sums them into one final smoothed curve.
This curve estimates the probability density of the data distribution.

Each kernel in kernel_list is a 1D array of y-values (PDF values) evaluated over x_axis â€” say, 100 values.
So if you have 25 data points, kernel_list is essentially a 2D structure like: shape: (25, 100)
This means:
â†’ For each x-value, sum the y-values from all 25 kernels at that x.
âœ… This is what gives you the KDE: A single curve that represents the total contribution at each x.
'''

# Plot figure
fig = plt.plot(x_axis,sum_of_kde,color='indianred')

# Add the initial rugplot
sns.rugplot(dataset,c = 'indianred')

# Get rid of y-tick marks
plt.yticks([])  #This removes the y-axis tick marks: Aesthetic choice â€” since you're focused on shape rather than actual density values.
# Set title
plt.suptitle("Sum of the Basis Functions");
```
