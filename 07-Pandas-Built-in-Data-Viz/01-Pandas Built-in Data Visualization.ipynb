# Plot Types

There are several plot types built-in to pandas, most of them statistical plots by nature:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)
___

Let's start going through them!

## Area

```python
# df2.shape[1]  #4
len(df2.columns)    #4
df2.columns #Index(['a', 'b', 'c', 'd'], dtype='object')
df2.index   #RangeIndex(start=0, stop=10, step=1)
```

```python
df2.plot(kind='area', alpha=0.4);
```

```python
df2.plot.area(alpha=0.4);
'''
What It Does
df2.plot.area():
Creates an area plot from the DataFrame df2.
Each column of the DataFrame is plotted as a filled area under a curve.
The x-axis is the index of df2.
The y-axis shows the values for each column.

alpha=0.4:
Sets the transparency level of the filled areas.
alpha ranges from 0 (completely transparent) to 1 (completely opaque).
So 0.4 means 40% opaque and 60% see-through — useful when areas overlap.
''';
```

## Line Plots

```python
df1.columns #Index(['A', 'B', 'C', 'D'], dtype='object')
```

```python
df1.plot.line(y='B',figsize=(12,3),lw=1);
# df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1); is not correct!
# The x parameter in df1.plot.line() should be a column name, not the index itself. If you want to plot the index on the x-axis (which is common), you don't need to specify x explicitly — it's automatically used.
'''
df1: Your DataFrame — it must contain a column named 'B'.
.plot.line(): This calls the line plot function (same as df1.plot(kind='line')).
y='B': Specifies that the data in column 'B' should be plotted on the y-axis. The x-axis will be the DataFrame's index by default.
figsize=(12, 3): Sets the size of the plot — 12 inches wide and 3 inches tall.
lw=1: Line width (thickness of the line) is set to 1.
'''
```

## Scatter Plots

```python
df1['C'].max()  #3.3424844555880395
df1['C'].min()  #-2.987971138961773
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm');  # We're now displaying three levels of information: the position of A vs B, and the color represents the value of C.
```

You can use c to color based off another column value
Use cmap to indicate colormap to use. 
For all the colormaps, check out: http://matplotlib.org/users/colormaps.html

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='rainbow');
```

Or use s to indicate size based off another column. s parameter needs to be an array, not just the name of a column:

```python
print(df1['C'].isna().sum())      # Check for NaNs
print((df1['C'] < 0).sum())       # Check for negative values
```

```python
df11 = df1[df1['C'] >= 0]            # Remove negative values
# df1['C'] = df1['C'].fillna(0)      # Or fill NaNs

df11.plot.scatter(x='A',y='B',s=df11['C']*200);
```

```python
df11.plot.scatter(
    x='A',
    y='B',
    c='C',           # color of fill
    cmap='coolwarm',
    s=df11['C']*100,           # size of circles
    edgecolors='black',  # color of boundary
    linewidths=1.5       # width of boundary
);
```

## BoxPlots

```python
df2.plot.box(); # Can also pass a by= argument for groupby
```

```python
import pandas as pd

df = pd.DataFrame({
    'group': ['A', 'A', 'B', 'B', 'B', 'A', 'A', 'B'],
    'value1': [10, 12, 14, 15, 16, 13, 11, 13],
    'value2': [5, 6, 7, 8, 7, 6, 5, 8]
})
df.plot.box(by='group')
'''
What it does:
It creates separate boxplots for each value column (value1, value2).
Each plot will be split by the unique values in the 'group' column (A and B in this case).
So you'll see side-by-side box plots per group.

Summary:
✅ Use by='column_name' to group your data by a categorical column.
🔍 Helpful for comparing distributions across groups.
''';
```

## Hexagonal Bin Plot

Useful for Bivariate Data, alternative to scatterplot:

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='BuPu')
'''
np.random.randn(1000, 2): Generates a 1000 x 2 array of random numbers from a standard normal distribution (mean = 0, std = 1). This will be your dataset.
columns=['a', 'b']:Names the two columns a and b.
df.plot.hexbin(...): Creates a hexbin plot — a 2D density plot that groups points into hexagonal bins. Each hexagon shows how many points fall into that region of the plot.

🔧 Parameters:
x='a', y='b': You're plotting column 'a' on the x-axis and column 'b' on the y-axis.
gridsize=25: Controls the number of hexagons (larger value → smaller hexagons → more detail).
cmap='Oranges': Color map used to indicate density — darker/oranger hexagons have more points.

📊 What the plot shows:
A visualization of how densely packed your data points are in different areas of the (a, b) coordinate space.
Useful for large datasets where scatter plots get cluttered.

When to use:
When you have a lot of overlapping points and want to see data density.
It’s a great alternative to scatter plots for big datasets.
''';
```
